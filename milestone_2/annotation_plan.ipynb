## Annotation examples

As discussed in the project proposal, the team will be annotating the central entity/subject of the news articles specifically pertaining to the **entertainment** category. Additionally, we will also annotate the "type" of the identified entity into following categories:

- **FILM**: If the central entity of the text is a movie, documentary, film-franchise, TV series etc., it's annotated as `type: FILM`. For example: movie reviews.

```python
example_df_3.loc[[0, 1, 2]]
```

- **PERFORMER**: If the central entity of the text is an actor/actress for any sort of film, it's annotated as `type: PERFORMER`. 

```python
example_df_3.loc[[20, 23, 93]]
```

- **FILM-CREW**: If the central entity of the text is any person or company/organization related to film other than the `PERFORMER`, it's annotated as `type: FILM-CREW`. Examples include directors, producers, movie studios, etc.

```python
example_df_3.loc[[45, 56, 70]]
```

- **OTHERS**: If the central entity of the text is anything other than a `FILM`, `PERFORMER` or a `FILM-CREW`, it's annotated as `type: OTHERS`.

```python
example_df_3.loc[[29, 37, 78]]
```

### Additional considerations

- If two entities are identified as central in the text, consider anyone of the two but not both. For example, if `A` and `B` are identified as central entities, annotate the central entity as either `A` or `B`.
- If an article seems to have no central entity, annotate it as `None`. In this case, the type of this entity will be `OTHERS`. Consider the following example:

```python
example_df_3['Text'][55]
```

In this case, we can see that the article asks the readers about their favorite movie. Since there's no central entity here, we can annotate it as `None`. Here is another example:

```python
example_df_3['Text'][78]
```

```python
example_df_3.loc[[55, 78]]
```

## Pilot Study

A small pilot study was conducted with two published tasks to verify if the project will be picked up at all. It got picked up in less than 10 minutes. However, the task was taking too long to be completed and the results were not promising. This could be attributed to two reasons:
- The instructions are unclear and need further work upon.
- The task is too complicated.

The latter should ideally not be the case since the task isn't that complicated. To correct the former, the instructions were re-written and another batch was published. The project got picked up again and though there was a minor improvement in the understanding of how the annotations were supposed to be done, the annotations were still off by a little. After rejecting two assignments however, the next proposed assignments were accurate and hence, approved. Considering this, the pilot study can be considered a success. Nonetheless, further improvements are required in communicating the annotation instructions. Else, alternatives besides Mechanical Turk would have to be considered after consultation with the instructors.
